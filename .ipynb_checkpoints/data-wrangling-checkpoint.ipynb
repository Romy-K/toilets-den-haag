## Combining Rotterdam data into 1 df
--> I ended up not using the Rotterdam data, so this is just in case I will continue the project at a later time. 

```python

urls = [
    "https://kaartlaag.rotterdam.nl/api/layers/15207.json",
    "https://kaartlaag.rotterdam.nl/api/layers/15210.json",
    "https://kaartlaag.rotterdam.nl/api/layers/15216.json",
    "https://kaartlaag.rotterdam.nl/api/layers/15219.json",
    "https://kaartlaag.rotterdam.nl/api/layers/15213.json",
]

all_rows = []

for url in urls:
    response = requests.get(url)
    data = response.json()
    for feat in data['features']:
        props = feat['properties']
        coords = feat['geometry']['coordinates']
        all_rows.append({**props, 'lon': coords[0], 'lat': coords[1]})

df = pd.DataFrame(all_rows)
df.to_csv("rotterdam_toilets.csv", index=False)
print("Saved combined dataset with", len(df), "rows")
```

## Create city_stats df and scenarios function 

```python
# City stats is where all toilet stats for a city (in this case, Den Haag) are stored. 
# It is the basis for (future) scenarios that add public toilets to see the economic effect. 

seniors_2024 = dh_seniors.loc[
    (dh_seniors['year'] == 2024),
    'seniors'
].values[0] 

avg_spend = leisure_survey.loc[
    (leisure_survey['age_group'] == 'senior'),
    'avg_spend_eur'
].values[0] 

city_stats = pd.DataFrame({
    "city": ['Den Haag'],
    "toilets": [67], # open and public toilets from dh_toilets 
    "land_area_km2": [82.44], # source: Wikipedia 
    "seniors": [seniors_2024],
    "avg_spend_per_outing": [avg_spend],
    "toilets_per_km2": [None],
    "ideal_toilets": [None],
    "coverage_ratio": [None],
    "cost_new_toilet": [200000], # at the high end of the costs I've seen
    "senior_outings_observed": [leisure_survey.loc[leisure_survey['age_group'] == 'seniors_dh', 'annual_outings'].values[0]],
    "senior_outings_ideal": [None],
    "missed_outings": [None],
    "missed_outing_value_eur": [None]
})

city_stats
```

```python
# Create function to calculate scenarios 

def sigmoid_response(coverage, midpoint=0.5, steepness=10):
    """Returns a value between 0 and 1 for a given coverage using an S-curve. 
    This makes for a more realistic behavioral simulation than a linear model."""
    return 1 / (1 + np.exp(-steepness * (coverage - midpoint)))

def add_city_scenario(df, toilets, base_stats, scenario_name, midpoint=0.5, steepness=10):
    """
    Creates economic scenarios based on the amount of toilets in the city, with the current 67 toilets as a baseline.
    Enter in a hypothetical amount of toilets, and you'll see how many senior outings are missed, how many euros spent 
    those outings represent, what the new toilets would cost, and what the net gain or loss is for the city.
    """
    # Behavioral assumption: 'sometimes' = 40% of the time; 'regularly' = 70% of the time
    missed_fraction = 0.233  # 11% skip 70% + 39% skip 40%

    # Static inputs from base_stats single-row DataFrame
    city = base_stats["city"].iat[0]
    current_toilets = base_stats["toilets"].iat[0]
    cost_per_toilet = base_stats["cost_new_toilet"].iat[0]
    land_area = base_stats["land_area_km2"].iat[0]
    avg_spend = base_stats["avg_spend_per_outing"].iat[0]
    observed_outings = base_stats["senior_outings_observed"].iat[0]

    # Coverage calculations
    ideal_toilets = land_area * 4
    coverage_current = current_toilets / ideal_toilets
    coverage_new = toilets / ideal_toilets

    # Calculate sigmoid values at baseline and ideal coverage
    sigmoid_baseline = sigmoid_response(coverage_current, midpoint, steepness)
    sigmoid_ideal = sigmoid_response(1.0, midpoint, steepness)  # Should be near 1

    # Raw sigmoid for new coverage
    relative_gain_raw = sigmoid_response(coverage_new, midpoint, steepness)

    # Rescale sigmoid so baseline coverage maps to zero gain (all missed outings),
    # ideal coverage maps to full gain (no missed outings)
    if relative_gain_raw <= sigmoid_baseline:
        relative_gain = 0.0
    elif relative_gain_raw >= sigmoid_ideal:
        relative_gain = 1.0
    else:
        relative_gain = (relative_gain_raw - sigmoid_baseline) / (sigmoid_ideal - sigmoid_baseline)

    # Outings calculations
    # Total outings if no one missed any outings (ideal)
    outings_ideal = observed_outings / (1 - missed_fraction)
    missed_total = outings_ideal - observed_outings

    # Remaining missed outings after improvement from added toilets
    missed_remaining = missed_total * (1 - relative_gain)

    # Value lost due to missed outings remaining, and value recovered by gain
    value_remaining = missed_remaining * avg_spend
    value_recovered = (missed_total - missed_remaining) * avg_spend

    # Cost & net gain
    cost_install = max(0, toilets - current_toilets) * cost_per_toilet
    net_gain = value_recovered - cost_install

    # Rounded values
    toilets_per_km2_rounded = round(toilets / land_area, 2)
    ideal_toilets_rounded = round(ideal_toilets)
    coverage_ratio_rounded = round(coverage_new, 3)

    # Assemble the row
    row = {
        "city": city,
        "scenario": scenario_name,
        "toilets": toilets,
        "toilets_per_km2": toilets_per_km2_rounded,
        "ideal_toilets": ideal_toilets_rounded,
        "coverage_ratio": coverage_ratio_rounded,
        "senior_outings_observed": observed_outings,
        "senior_outings_ideal": outings_ideal,
        "missed_outings": missed_remaining,
        "missed_outing_value_eur": value_remaining,
        "cost_new_toilets_eur": cost_install,
        "net_gain_eur": net_gain,
        "relative_gain": relative_gain,
        "sigmoid_midpoint": midpoint,
        "sigmoid_steepness": steepness
    }

    return pd.concat([df, pd.DataFrame([row])], ignore_index=True)
```

```python
city_scenarios = pd.DataFrame()

dh_base = city_stats[city_stats["city"] == "Den Haag"]

city_scenarios = add_city_scenario(city_scenarios, 67, dh_base, "baseline")
city_scenarios = add_city_scenario(city_scenarios, 68, dh_base, "just the one")
city_scenarios = add_city_scenario(city_scenarios, 100, dh_base, "strong")
city_scenarios = add_city_scenario(city_scenarios, 330, dh_base, "ideal")

city_scenarios
```

```python
# Get all scenarios
all_scenario_dfs = [] # This list will collect each single-row DataFrame scenario

# Loop from 67 up to and including 330
for t in range(67, 331, 1): # Range to include 330
    # Pass an EMPTY DataFrame as the first argument so it only returns the new row
    new_scenario_df = add_city_scenario(pd.DataFrame(), t, dh_base, f"{t} toilets")
    all_scenario_dfs.append(new_scenario_df)

# Concatenate all the single-row DataFrames into one large DataFrame
city_scenarios_all = pd.concat(all_scenario_dfs, ignore_index=True)

print(f"Total rows in city_scenarios_all: {len(city_scenarios_all)}")
print(f"Unique toilet counts: {city_scenarios_all['toilets'].nunique()}")
print(f"Min toilet count: {city_scenarios_all['toilets'].min()}")
print(f"Max toilet count: {city_scenarios_all['toilets'].max()}")
```

## Editing dh_seniors to include 2034 for my 10-year simulation 

```python
from sklearn.linear_model import LinearRegression

# Using linear regression to calculate (estimate) 2034 numbers
years = dh_seniors["year"].values.reshape(-1, 1)
total_residents = dh_seniors["total_residents"].values
seniors = dh_seniors["seniors"].values

# Fit linear regressions
model_residents = LinearRegression().fit(years, total_residents)
model_seniors = LinearRegression().fit(years, seniors)
```

```python
year_to_predict = np.array([[2034]])
predicted_total_residents = model_residents.predict(year_to_predict)[0]
predicted_seniors = model_seniors.predict(year_to_predict)[0]

# Assume % stays constant
seniors_percent = round((predicted_seniors / predicted_total_residents) * 100, 1)
```
